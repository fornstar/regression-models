```python
# Configuration & Setup
import numpy as np
import os

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
# Create a randomized linear dataset 
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
#let's plot it to see what it looks like
plt.plot(X,y,"b.") 

# add x0 = 1 to each instance
X_b = np.c_[np.ones((100, 1)), X]  

#create an array to store two coordinates for regression line 
X_new = np.array([[0], [2]])

# add x0 = 1 to each instance
X_new_b = np.c_[np.ones((2, 1)), X_new]  
```

```python
## Construct the gradient descent
```

```python
# eta is the hyperparameter that controls the size of the downhill step
eta = 0.1
n_iterations = 1000 #number of training runs
m = 100 #sample size
theta = np.random.randn(2,1) #initialise theta

for iteration in range(n_iterations):
    
    #calculate the gradient using the partial derivative of the cost function MSE
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y) 
    
    #perform the descent
    theta = theta - eta * gradients #calculate the new value of theta
    
    #keep track of y
    y_predict = X_new_b.dot(theta)
    plt.plot(X_new, y_predict)
    
plt.xlabel("$x_1$", fontsize=18)
plt.axis([0, 2, 0, 15])
plt.title(r"$\eta = {}$".format(eta), fontsize=16)

#plot the orginal dataset
plt.plot(X, y, "b.")
```
