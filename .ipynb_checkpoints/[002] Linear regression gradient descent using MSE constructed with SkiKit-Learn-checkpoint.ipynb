```python
# Configuration & Setup
import numpy as np
import os

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
# Create a randomized linear dataset 
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
#let's plot it to see what it looks like
plt.plot(X,y,"b.") 

# add x0 = 1 to each instance
X_b = np.c_[np.ones((100, 1)), X]  

#create an array to store two coordinates for regression line 
#x points are limit of our plot
X_new = np.array([[0], [2]])
```

```python
## Construct the gradient descent
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_

#Compute the y values that correspond to the x=0 and x=2 values (limits of the plot)
y_predict = lin_reg.predict(X_new)

plt.plot(X,y,"b.")

#plot the regression line based on current value of theta
plt.plot(X_new, y_predict)
```
